```python
import pandas as pd
import numpy as np
```

```python
stations = pd.read_csv("stations.csv")
price = pd.read_csv("price.csv")
occupancy = pd.read_csv("occupancy.csv")
volume = pd.read_csv("volume.csv")
duration = pd.read_csv("duration.csv")
distance = pd.read_csv("distance.csv")
time_df = pd.read_csv("time.csv")
information = pd.read_csv("information.csv")
adj = pd.read_csv("adj.csv")
```

```python
time_df["datetime"] = pd.to_datetime({
    "year": time_df["year"],
    "month": time_df["month"],
    "day": time_df["day"],
    "hour": time_df["hour"],
    "minute": time_df["minute"],
    "second": time_df["second"]
})
```

```python
time_df.head()
```

```python
time_df["timestamp"] = range(
    1,
    len(time_df)+1
)
```

```python
time_df.head()
```

```python
price_long = price.melt(
    id_vars=["timestamp"],
    var_name="station_id",
    value_name="price"
)
```

```python
price_long.head()
```

```python
occupancy_long = occupancy.melt(
    id_vars=["timestamp"],
    var_name="station_id",
    value_name="occupancy"
)
```

```python
occupancy_long.head()
```

```python
volume_long = volume.melt(
    id_vars=["timestamp"],
    var_name="station_id",
    value_name="volume"
)
```

```python
duration_long = duration.melt(
    id_vars=["timestamp"],
    var_name="station_id",
    value_name="duration"
)
```

```python
print(price_long.shape)
print(occupancy_long.shape)
print(volume_long.shape)
print(duration_long.shape)
```

```python
print(price_long.shape)
print(price_long.head())

print(occupancy_long.shape)
print(occupancy_long.head())
```

```python
volume_long = volume.melt(
    id_vars=["timestamp"],
    var_name="station_id",
    value_name="volume"
)

duration_long = duration.melt(
    id_vars=["timestamp"],
    var_name="station_id",
    value_name="duration"
)
```

```python
price_long["station_id"] = price_long["station_id"].astype(int)
occupancy_long["station_id"] = occupancy_long["station_id"].astype(int)
volume_long["station_id"] = volume_long["station_id"].astype(int)
duration_long["station_id"] = duration_long["station_id"].astype(int)
```

```python
master = price_long.merge(
    occupancy_long,
    on=["timestamp", "station_id"]
)

master = master.merge(
    volume_long,
    on=["timestamp", "station_id"]
)

master = master.merge(
    duration_long,
    on=["timestamp", "station_id"]
)
```

```python
print(master.shape)
master.head()
```

```python
time_df["timestamp"] = range(
    1,
    len(time_df)+1
)
```

```python
master = master.merge(
    time_df,
    on="timestamp",
    how="left"
)
```

```python
master.head()
```
