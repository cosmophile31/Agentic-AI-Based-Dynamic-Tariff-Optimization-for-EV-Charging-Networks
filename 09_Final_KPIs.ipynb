```python
import pandas as pd

demand_metrics = pd.read_csv(
    "demand_metrics.csv"
)

tariff_metrics = pd.read_csv(
    "tariff_metrics.csv"
)

monitoring_metrics = pd.read_csv(
    "monitoring_metrics.csv"
)

acn_metrics = pd.read_csv(
    "acn_metrics.csv"
)
```

```python
demand_metrics["Agent"] = "Demand Prediction"

tariff_metrics["Agent"] = "Tariff Pricing"

monitoring_metrics["Agent"] = "Monitoring"

acn_metrics["Agent"] = "Monitoring / ACN"
```

```python
utilization_table = pd.DataFrame({

    "Utilization Metric":[

        "Baseline Utilization Rate",

        "Post-Pricing Utilization Rate",

        "Utilization Improvement (%)"
    ],

    "Value":[

        0.2802482048319062,      # util_before

        0.2800705818069866,      # util_after

        -0.0633800672955042      # utilization_improvement
    ]
})

utilization_table
```

```python
final_kpis = pd.concat(

    [
        demand_metrics,
        tariff_metrics,
        monitoring_metrics,
        acn_metrics
    ],

    ignore_index=True
)

final_kpis
```

# Assumptions

1. A fixed tariff of ₹15/kWh was used as the baseline pricing benchmark.

2. Customer response to pricing changes is not directly available in the datasets. Therefore, elasticity assumptions were introduced:

   * Discount pricing increases charging demand by 20%.
   * Surge pricing reduces charging demand by 2%.

3. Queue length is not explicitly available in UrbanEV; therefore, queue_proxy was used as a waiting-time surrogate.

4. ACN session energy delivered (kWhDelivered) was used as the primary indicator for pricing optimization and revenue evaluation.


```python
utilization_table.to_csv(
    "utilization_metrics.csv",
    index=False
)

print("Utilization metrics saved successfully.")
```
